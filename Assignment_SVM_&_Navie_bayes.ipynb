1. What is a Support Vector Machine (SVM)?

Ans:-
A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification, regression, and outlier detection. It is especially powerful for classification tasks.


 2. What is the difference between Hard Margin and Soft Margin SVM?

 Ans:- A Hard Margin SVM is used when the data is perfectly linearly separable and requires that no training examples are misclassified. It finds a hyperplane that separates the classes with the maximum possible margin, but it cannot handle noise or outliers because even a single misclassified point makes the model invalid. In contrast, a Soft Margin SVM allows some misclassification by introducing slack variables, making it more flexible and suitable for real-world datasets that contain noise or overlapping classes. The flexibility is controlled by the parameter C, which balances margin maximization and error minimization. Soft Margin SVM is therefore more commonly used, as it handles imperfect data more effectively than Hard Margin SVM.


3. What is the mathematical intuition behind SVM?

Ans:- The mathematical intuition behind SVM is that it tries to find a hyperplane that separates classes with the largest possible margin. The margin is the distance between the hyperplane and the closest training points from each class, called support vectors. Mathematically, SVM chooses the hyperplane that maximizes this margin while still correctly classifying (or nearly classifying) the data. For linearly separable data, this becomes an optimization problem where we minimize the norm of the weight vector ‖w‖² subject to all points being correctly classified. A smaller ‖w‖ means a larger margin, which improves generalization. When the data is not perfectly separable, slack variables allow some points to violate the margin, and the optimization balances margin maximization with misclassification penalties controlled by the parameter C. In nonlinear cases, the kernel trick maps the data into a higher-dimensional space where a linear boundary can be found, while avoiding explicitly computing the high-dimensional features. Thus, SVM mathematically seeks the simplest, widest-margin separator that best generalizes to unseen data.


4. What is the role of Lagrange Multipliers in SVM?

Ans:- In SVM, Lagrange Multipliers are used to convert the constrained optimization problem of maximizing the margin into a form that can be solved more efficiently. Instead of directly optimizing the hyperplane under inequality constraints, SVM uses the Lagrangian to incorporate these constraints into the objective function. Solving the dual problem with Lagrange Multipliers identifies which training points become support vectors, because only points with non-zero multipliers lie on or violate the margin. This makes the computation more efficient, allows the kernel trick to be applied in nonlinear SVMs, and provides a clear mathematical way to focus the model on only the most important data points.


5. What are Support Vectors in SVM?

Ans:-
Support Vectors are the specific training data points that lie closest to the decision boundary or margin in a Support Vector Machine. These points are critical because they directly determine the position and orientation of the separating hyperplane. Only the Support Vectors influence the final model; removing any other points would not change the decision boundary. They are also the points with non-zero Lagrange multipliers, meaning they either lie exactly on the margin or violate it in the soft-margin case. In essence, Support Vectors are the key examples that define the classifier and give SVM its robustness and generalization ability.


6. What is a Support Vector Classifier (SVC)?

Ans:- A Support Vector Classifier (SVC) is the practical implementation of the Support Vector Machine for classification tasks. It works by finding a decision boundary that separates data points of different classes with the maximum possible margin. SVC can use either a linear boundary or nonlinear boundaries through kernel functions, which allow the data to be mapped into higher-dimensional spaces where separation is easier. It focuses only on the support vectors—the most informative training points near the boundary—to determine the classifier. By balancing margin width and classification errors through the parameter C, SVC provides a flexible and powerful method for handling both clean and noisy datasets.


7.  What is a Support Vector Regressor (SVR)?

Ans:- A Support Vector Regressor (SVR) is the regression counterpart of a Support Vector Machine, used to predict continuous values instead of class labels. SVR works by finding a function that deviates from the actual target values by at most a specified margin, called epsilon (ε), while keeping the model as flat as possible. Only data points that lie outside this epsilon-insensitive margin, known as support vectors, influence the final regression function. SVR can also use kernel functions to handle nonlinear relationships, and it balances the trade-off between margin width and prediction errors using a regularization parameter C. This makes SVR robust, flexible, and effective for regression tasks with both linear and nonlinear data.


8.  What is the Kernel Trick in SVM?

Ans:- The Kernel Trick in SVM is a mathematical technique that allows the algorithm to handle nonlinearly separable data by implicitly mapping it into a higher-dimensional space without explicitly computing the new coordinates. This is done using kernel functions, which calculate the inner product of data points in the transformed space directly from the original features. Common kernels include linear, polynomial, radial basis function (RBF), and sigmoid. By using the Kernel Trick, SVM can find a linear hyperplane in the higher-dimensional space that corresponds to a nonlinear decision boundary in the original input space, making it a powerful tool for complex classification and regression problems.


9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

Ans:- The Linear Kernel is the simplest kernel, suitable for linearly separable data, and computes the inner product of two vectors. It is fast, requires fewer parameters, and works well when the number of features is large. The Polynomial Kernel maps data into a higher-dimensional space using polynomial functions of the original features, allowing it to model nonlinear relationships; its flexibility depends on the degree of the polynomial, but higher degrees can lead to overfitting. The RBF (Radial Basis Function) Kernel maps data into an infinite-dimensional space and measures similarity based on distance between points, making it highly effective for complex, nonlinear datasets. It is the most popular kernel due to its ability to handle intricate decision boundaries, though it requires careful tuning of the gamma parameter to control influence and avoid overfitting.


10. What is the effect of the C parameter in SVM?

Ans:- The C parameter in SVM controls the trade-off between maximizing the margin and minimizing classification errors. A large C assigns a higher penalty to misclassified points, forcing the model to fit the training data more strictly and resulting in a narrower margin, which may lead to overfitting. A small C allows more misclassifications, creating a wider margin and making the model more flexible and generalizable to unseen data. In essence, C determines how much the SVM prioritizes correct classification of training points versus a larger, more robust margin.


11. What is the role of the Gamma parameter in RBF Kernel SVM?

Ans:- In an RBF (Radial Basis Function) Kernel SVM, the gamma (γ) parameter defines the influence of a single training point on the decision boundary. A small gamma value means that each point has a wide influence, resulting in a smoother, simpler decision boundary that may underfit the data. A large gamma value makes each point’s influence very narrow, allowing the model to capture fine details and complex patterns, but it may overfit the training data. Essentially, gamma controls the “reach” of individual support vectors and directly affects the flexibility and complexity of the model.


12. What is the Naïve Bayes classifier, and why is it called "Naïve"?

Ans:-  The Naïve Bayes classifier is a probabilistic machine learning algorithm used for classification, based on Bayes’ Theorem, which calculates the probability of a class given the observed features. It is called “naïve” because it assumes that all features are conditionally independent of each other given the class, an assumption that is rarely true in real-world data. Despite this strong and often unrealistic assumption, Naïve Bayes performs surprisingly well in many applications such as text classification, spam detection, and sentiment analysis, because it captures the overall patterns of probability effectively.


13.  What is Bayes’ Theorem?

Ans:- Bayes’ Theorem is a fundamental principle in probability theory that describes how to update the probability of a hypothesis based on new evidence. Mathematically, it states that P(A|B) = [P(B|A) × P(A)] / P(B), where P(A|B) is the probability of event A given that B has occurred, P(B|A) is the probability of event B given A, P(A) is the prior probability of A, and P(B) is the prior probability of B. Bayes’ Theorem provides a way to combine prior knowledge with observed data to make probabilistic inferences and is the foundation of probabilistic classifiers like Naïve Bayes.


14.  Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.

Ans:- Gaussian Naïve Bayes assumes that the features are continuous and follow a normal (Gaussian) distribution, making it suitable for real-valued data such as heights, weights, or temperatures. Multinomial Naïve Bayes is designed for discrete count data, commonly used in text classification tasks where features represent word counts or frequencies. Bernoulli Naïve Bayes is used for binary/boolean features, where each feature indicates the presence or absence of an attribute, such as whether a word appears in a document. In summary, Gaussian is for continuous data, Multinomial for count-based discrete data, and Bernoulli for binary data, with each variant applying Bayes’ Theorem under its respective assumptions.


15.  When should you use Gaussian Naïve Bayes over other variants.

Ans:-  Gaussian Naïve Bayes should be used when the features in your dataset are continuous numerical values and are approximately normally distributed. It is suitable for tasks like predicting heights, weights, temperatures, or other measurements where the assumption of a Gaussian distribution is reasonable. Unlike Multinomial or Bernoulli variants, which are designed for discrete counts or binary features, Gaussian Naïve Bayes handles real-valued data directly and calculates probabilities using the mean and variance of each feature, making it the preferred choice for continuous feature datasets.


16.  What are the key assumptions made by Naïve Bayes?

Ans:- Naïve Bayes makes two key assumptions: first, it assumes that all features are conditionally independent given the class label, meaning the presence or value of one feature does not affect another within the same class. Second, it assumes that the features follow a specific probability distribution depending on the variant used—Gaussian for continuous features, Multinomial for discrete counts, and Bernoulli for binary features. Despite these strong and often unrealistic assumptions, Naïve Bayes works well in practice because it effectively captures the overall probabilistic patterns in the data.

17.  What are the advantages and disadvantages of Naïve Bayes?

Ans:-

Advantages: Naïve Bayes is simple, fast, and easy to implement, making it suitable for large datasets. It works well with high-dimensional data and performs surprisingly well even when the independence assumption is violated. It is also effective for text classification, spam detection, and real-time predictions.

 Disadvantages: Its main limitation is the strong assumption of feature independence, which may not hold in real-world data and can reduce accuracy. It also struggles with features that have zero probability in the training data unless smoothing techniques are applied, and it may be less flexible than more complex models for capturing intricate relationships.


18. Why is Naïve Bayes a good choice for text classification?

Ans:- Naïve Bayes is well-suited for text classification because it handles high-dimensional feature spaces efficiently, such as the large number of words in a vocabulary. It works effectively with sparse data, where most features (words) are absent in a given document. The algorithm also models word occurrence probabilities naturally, making variants like Multinomial Naïve Bayes ideal for tasks such as spam detection, sentiment analysis, and topic categorization. Despite its independence assumption, Naïve Bayes often performs very well on text data because the presence or absence of individual words carries strong predictive power.


19. Compare SVM and Naïve Bayes for classification tasks.

Ans:- Support Vector Machine (SVM) and Naïve Bayes are both popular classification algorithms, but they differ in approach and applicability. SVM is a discriminative model that finds an optimal hyperplane to separate classes, works well with high-dimensional and non-linear data using kernel functions, and is robust to overfitting when properly regularized. Naïve Bayes is a probabilistic model based on Bayes’ Theorem, assumes feature independence, and is particularly efficient for large, sparse datasets such as text classification. SVM generally provides higher accuracy on complex datasets with overlapping features, while Naïve Bayes is faster, easier to implement, and performs well when the independence assumption roughly holds. In summary, SVM is better for complex or non-linear patterns, whereas Naïve Bayes is ideal for high-dimensional, sparse, or text-based data.


20.  How does Laplace Smoothing help in Naïve Bayes?

Ans:-  Laplace smoothing is used in Naïve Bayes to handle the problem of zero probabilities that occur when a feature value does not appear in the training data for a given class. Without smoothing, a zero probability would make the entire class probability zero, regardless of other evidence. Laplace smoothing adds a small constant (usually 1) to all feature counts, ensuring that no probability is ever zero. This allows the model to make more robust predictions, improves generalization to unseen data, and prevents overfitting to the training dataset.


```python
# Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling (important for SVM performance)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create and train the SVM classifier
svm_clf = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_clf.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_clf.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM classifier on Iris dataset: {accuracy:.4f}")
```

```python
# 22. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train SVM with Linear kernel
svm_linear = SVC(kernel='linear', C=1.0)
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)

# Train SVM with RBF kernel
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

# Print accuracies
print(f"Accuracy of Linear SVM: {accuracy_linear:.4f}")
print(f"Accuracy of RBF SVM: {accuracy_rbf:.4f}")

# Compare which kernel performs better
if accuracy_linear > accuracy_rbf:
    print("Linear kernel performs better on this dataset.")
elif accuracy_rbf > accuracy_linear:
    print("RBF kernel performs better on this dataset.")
else:
    print("Both kernels perform equally well on this dataset.")
```

```python
#23.Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using MeanSquared Error (MSE)

# Import necessary libraries
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

# Load California Housing dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Scale features
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

# Scale target variable
scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()

# Initialize and train SVR with RBF kernel
svr_model = SVR(kernel='rbf', C=50, gamma=0.05, epsilon=0.1)
svr_model.fit(X_train_scaled, y_train_scaled)

# Predict on test data
y_pred_scaled = svr_model.predict(X_test_scaled)

# Inverse transform predictions to original scale
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))

# Evaluate performance using Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"SVR Mean Squared Error: {mse:.4f}")
```

```python
#24.Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load a 2D dataset (Iris dataset, using only first two features for visualization)
iris = datasets.load_iris()
X = iris.data[:, :2]  # only first two features for 2D plot
y = iris.target

# For simplicity, let's use only two classes (setosa vs versicolor)
X = X[y != 2]
y = y[y != 2]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train SVM with Polynomial kernel
poly_svm = SVC(kernel='poly', degree=3, C=1.0, gamma='scale')
poly_svm.fit(X_scaled, y)

# Create a meshgrid for plotting decision boundary
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Predict on meshgrid points
Z = poly_svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=50, cmap=plt.cm.Paired, edgecolors='k')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.title('SVM with Polynomial Kernel (degree=3)')
plt.show()
```

```python
#25. Write a Python program to train a Gaussian Naïve Bayes classifier on the Breast Cancer dataset and evaluate accuracy.

# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Feature scaling (optional but can improve performance)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train Gaussian Naïve Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = gnb.predict(X_test_scaled)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Gaussian Naïve Bayes on Breast Cancer dataset: {accuracy:.4f}")

# Optional: detailed evaluation
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
#26.  Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

# Import necessary libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the 20 Newsgroups dataset
categories = ['alt.atheism', 'comp.graphics', 'sci.med', 'soc.religion.christian']  # optional subset
newsgroups = fetch_20newsgroups(subset='all', categories=categories, shuffle=True, random_state=42)

X = newsgroups.data  # text data
y = newsgroups.target  # labels

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Convert text data to feature vectors
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
X_test_counts = count_vect.transform(X_test)

# Optionally use TF-IDF transformation
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_test_tfidf = tfidf_transformer.transform(X_test_counts)

# Train Multinomial Naïve Bayes classifier
mnb = MultinomialNB()
mnb.fit(X_train_tfidf, y_train)

# Predict on the test set
y_pred = mnb.predict(X_test_tfidf)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Multinomial Naïve Bayes: {accuracy:.4f}")

# Optional: detailed evaluation
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=newsgroups.target_names))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
#27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load 2D dataset (Iris dataset, first two features)
iris = datasets.load_iris()
X = iris.data[:, :2]  # first two features
y = iris.target

# For simplicity, use only two classes (setosa vs versicolor)
X = X[y != 2]
y = y[y != 2]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define different C values to compare
C_values = [0.1, 1, 10, 100]

# Create meshgrid for plotting
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Plot decision boundaries for each C
plt.figure(figsize=(15, 10))
for i, C in enumerate(C_values, 1):
    svm = SVC(kernel='linear', C=C)
    svm.fit(X_scaled, y)

    # Predict on meshgrid
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot
    plt.subplot(2, 2, i)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=50, cmap=plt.cm.Paired, edgecolors='k')
    plt.title(f'SVM Decision Boundary (C={C})')
    plt.xlabel('Feature 1 (scaled)')
    plt.ylabel('Feature 2 (scaled)')

plt.tight_layout()
plt.show()
```

```python
#30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ------------------------
# 1. Gaussian Naïve Bayes (no smoothing)
# ------------------------
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
print(f"Accuracy of Gaussian Naïve Bayes: {accuracy_gnb:.4f}")

# ------------------------
# 2. Multinomial Naïve Bayes with Laplace smoothing
# ------------------------
# Convert continuous features to discrete bins
kbd = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
X_train_binned = kbd.fit_transform(X_train)
X_test_binned = kbd.transform(X_test)

# Train MultinomialNB with Laplace smoothing (alpha=1)
mnb = MultinomialNB(alpha=1.0)
mnb.fit(X_train_binned, y_train)
y_pred_mnb = mnb.predict(X_test_binned)
accuracy_mnb = accuracy_score(y_test, y_pred_mnb)
print(f"Accuracy of Multinomial Naïve Bayes with Laplace smoothing: {accuracy_mnb:.4f}")
```

```python
#31.Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel).

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define SVM model
svm = SVC()

# Define hyperparameter grid for tuning
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly']
}

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Train model with hyperparameter tuning
grid_search.fit(X_train_scaled, y_train)

# Best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)

# Predict on test set using the best model
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test_scaled)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of best SVM model: {accuracy:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#33. Write a Python program to implement a Naïve Bayes classifier for spam detection using email data.

# Import necessary libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Simulate spam/ham using categories from 20 Newsgroups
categories = ['rec.sport.hockey', 'sci.crypt']  # 'sci.crypt' as spam, 'rec.sport.hockey' as ham
emails = fetch_20newsgroups(subset='all', categories=categories, shuffle=True, random_state=42)

X = emails.data  # email text
y = emails.target  # labels (0: ham, 1: spam)

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Convert text data to numeric features
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
X_test_counts = count_vect.transform(X_test)

# Transform counts to TF-IDF
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_test_tfidf = tfidf_transformer.transform(X_test_counts)

# Train Multinomial Naïve Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)

# Predict on the test set
y_pred = nb_classifier.predict(X_test_tfidf)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Spam detection accuracy: {accuracy:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=['ham', 'spam']))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

```

```python
#35.Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results.

# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, classification_report

# Load the dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ------------------------
# 1. Train Naïve Bayes without feature selection
# ------------------------
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_full = gnb.predict(X_test)
accuracy_full = accuracy_score(y_test, y_pred_full)
print(f"Accuracy without feature selection: {accuracy_full:.4f}")
print("Classification Report without feature selection:\n", classification_report(y_test, y_pred_full))

# ------------------------
# 2. Perform feature selection (Select top 10 features using chi-square)
# ------------------------
# Chi-square requires non-negative features, so scale to [0,1]
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Select top k features
selector = SelectKBest(score_func=chi2, k=10)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Train Naïve Bayes on selected features
gnb_selected = GaussianNB()
gnb_selected.fit(X_train_selected, y_train)
y_pred_selected = gnb_selected.predict(X_test_selected)
accuracy_selected = accuracy_score(y_test, y_pred_selected)
print(f"\nAccuracy with feature selection (top 10 features): {accuracy_selected:.4f}")
print("Classification Report with feature selection:\n", classification_report(y_test, y_pred_selected))
```

```python
#43. Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn.

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
class_names = iris.target_names

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train SVM classifier
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Predict on test set
y_pred = svm_model.predict(X_test_scaled)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Visualize confusion matrix using Seaborn
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - SVM Classifier')
plt.show()
```

```python
#44. Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE.

# Import necessary libraries
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# Generate synthetic regression dataset
X, y = make_regression(n_samples=500, n_features=5, noise=15, random_state=42)

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train SVR model
svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(X_train_scaled, y_train)

# Predict on test set
y_pred = svr_model.predict(X_test_scaled)

# Evaluate using Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.4f}")

# Optional: visualize actual vs predicted values
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('SVR: Actual vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()

```

```python
#45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.

# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale features (optional for GaussianNB)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Gaussian Naïve Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train_scaled, y_train)

# Predict probabilities for the positive class
y_prob = gnb.predict_proba(X_test_scaled)[:, 1]

# Compute ROC-AUC score
roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.4f}")

# Optional: Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Gaussian Naïve Bayes')
plt.legend()
plt.show()
```

```python
#46.Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

# Import necessary libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train SVM classifier with probability estimates
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', probability=True, random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Predict probabilities for positive class
y_prob = svm_model.predict_proba(X_test_scaled)[:, 1]

# Compute Precision-Recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
avg_precision = average_precision_score(y_test, y_prob)

# Plot Precision-Recall curve
plt.figure(figsize=(6,5))
plt.plot(recall, precision, color='blue', label=f'AP = {avg_precision:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve - SVM Classifier')
plt.legend()
plt.grid(True)
plt.show()
```
